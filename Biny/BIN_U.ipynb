```julia
using Random
using BenchmarkTools
using StatsBase
```

```julia
function BIN_U_1(x)
    return countmap(clamp.(ceil.(Int, 5 .* x), 1, 5))
end
```

```julia
function BIN_U_2(x)
    Bins = zeros(Int, 5)

     @inbounds for i in eachindex(x)
        b = Int(floor(5*x[i])) + 1
        Bins[b] += 1
    end
    return Bins
end
```

```julia
function BIN_U_3(x)
    Bins = zeros(Int, 5)
    @inbounds  for i in eachindex(x)
        b = trunc(Int, 5*x[i]) + 1
        Bins[b] += 1
    end
    return Bins
end
```

```julia
function BIN_U_4(x)
    Bins = zeros(Int, 5)
    @inbounds for i in eachindex(x)
        Bins[trunc(Int, 5 * x[i]) + 1] += 1
    end
    return Bins
end
```

```julia
function BIN_U_5(x)
    Bins = zeros(Int, 5)
    @inbounds @simd for i in eachindex(x)
        Bins[trunc(Int, 5 * x[i]) + 1] += 1
    end
    return Bins
end
```

```julia
function BIN_U_4_zmienne(x, bins)
    l = length(bins)
    @inbounds @simd for i in eachindex(x)
        bins[trunc(Int, l * x[i]) + 1] += 1
    end
    return bins
end
```

```julia
using BenchmarkTools

x = rand(10^6)
z = zeros(Int, 5)
@btime BIN_U_1($x)
GC.gc()
@btime BIN_U_2($x)
GC.gc()
@btime BIN_U_3($x)
GC.gc()
@btime BIN_U_4($x)
GC.gc()
@btime BIN_U_4_zmienne($x, $z)
GC.gc()
@btime BIN_U_5($x)

```
