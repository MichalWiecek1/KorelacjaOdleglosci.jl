```julia
# Nie widzę innej metody, próbowałem to jakoś transformować ale gubiły się ogony
function BIN_N(x)
    Bins = zeros(Int, 6)

    @inbounds for i in eachindex(x)
        xi = x[i]

        if xi < -2
            b = 1
        elseif xi < -1
            b = 2
        elseif xi < 0
            b = 3
        elseif xi < 1
            b = 4
        elseif xi < 2
            b = 5
        else
            b = 6
        end

        Bins[b] += 1
    end

    return Bins
end
```

```julia
# Trochę szybciej, mniej sprawdza
function BIN_N_bisekcja(x)
    Bins = zeros(Int, 6)

    @inbounds for i in eachindex(x)
        xi = x[i]

        if xi < 0
            if xi < -1
                if xi < -2
                    b = 1
                else
                    b = 2
                end
            else
                b = 3
            end
        else
            if xi < 1
                b = 4
            elseif xi < 2
                b = 5
            else
                b = 6
            end
        end

        Bins[b] += 1
    end

    return Bins
end
```
